```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import nltk
import spacy
import re

import warnings
warnings.filterwarnings(action="ignore")
```

```python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'cornell_movie_dialogs'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

dialogs_df = pd.read_sql_query('select * from dialogs',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()


dialogs_df.head(10)
```

```python
nlp = spacy.load('en', disable=['parser', 'ner'])
nlp.add_pipe(nlp.create_pipe('sentencizer'))
nlp.max_length = 20000000
dialog = " ".join(dialogs_df.dialogs)
dialog_doc = nlp(dialog)
```

```python
dialog_without_stopwords = [token for token in dialog_doc if not token.is_stop]
```

```python
dialog_without_stopwords
```

```python
dialog_lemmas = []
for token in dialog_without_stopwords:
    if not token.is_punct:
        dialog_lemmas.append(token.lemma_)
```
