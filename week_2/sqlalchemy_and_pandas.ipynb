```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import psycopg2
from sqlalchemy import create_engine
```

```
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'employees'
```

```
sql = '''
SELECT emp.empno, emp.ename, emp.job,
    emp.hiredate, emp.sal::decimal::money,
    mgr.empno AS mgr, mgr.ename AS mname,
    dept.deptno, dept.dname,
    loc.locno, loc.cty, loc."state", loc.zip
FROM emp
LEFT OUTER JOIN emp AS mgr
	ON emp.mgr = mgr.empno
INNER JOIN dept
	ON emp.deptno = dept.deptno
INNER JOIN loc
	ON dept.locno = loc.locno
ORDER BY 1
'''

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

# Since empno is a primary key for each employee, it works great as an index
df = pd.read_sql_query(sql,con=engine, index_col='empno')

# no need for an open connection, as we're only doing a single query
engine.dispose()

df.head(10)
```
