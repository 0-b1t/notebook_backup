```python
import numpy as np
import pandas as pd
import scipy
import plotly.express as px
from datetime import datetime
import datetime as dt
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import pacf

%matplotlib inline
```

```python
df = pd.read_csv("data/time_series_covid19_confirmed_US.csv")
df.head()
```

```python
data = df[df["Admin2"] == "San Diego"]
```

```python
data = data.drop(
    [
        "UID",
        "iso2",
        "iso3",
        "code3",
        "FIPS",
        "Admin2",
        "Province_State",
        "Country_Region",
        "Lat",
        "Long_",
        "Combined_Key",
    ],
    1,
)
```

```python
data=data.T.reset_index()
```

```python
data.columns = ["date", "cases"]
```

```python
data["date"] = pd.to_datetime(data["date"])
```

```python
def ilinechart(df, x, y, groups=None, title=""):
    fig = px.line(
        df, x=x, y=y, color=groups, title=title, template="none"
    )  # .update(layout=dict(title=dict(x=0.5)))

    fig.show()
```

```python
ilinechart(data, "date", "cases")
```

```python
data["shift"] = data["cases"].shift()
data["diff"] = data["cases"] - data["shift"]
```

```python
ilinechart(data, "date", "diff")
```
